```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

import os, json
```

```python
df = pd.read_csv('Data/basics.csv')
df.info()
df.head()
```

```python
df['genres_split'] = df['genres'].str.split(',')
df
```

## Getting a List of Unique Genres

```python
exploded_genres = df.explode('genres_split')
exploded_genres 
```

```python
unique_genres = sorted(exploded_genres['genres_split'].dropna().unique())
unique_genres
```

## Create a new title_genres table

```python
# Create a new title_genres table
## tconst and genre_id to match SQL 'title_genres' table
### save just tconst and genrest_split as new df 
title_genres = exploded_genres[['tconst', 'genres_split']].copy()
title_genres.head()
```

## Create a genre mapper dictionary to replace string genres with integers 

```python
# making the genre mapper dictionary 
## genre strings will be keys and genre_ids as the values 
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

## Replace the string genres in title_genres with the new integer ids 

```python
# make new integer genre_id and drop string genres 
df['genre_id'] = df['genres_split'].replace(genre_map)
df = df.drop(columns = 'genres_split')
```

## Convert the genre map dictionary into a dataframe 

```python
# manually make dataframe with named cols from the .keys and .values 
genre_lookup = pd.DataFrame({'Genre_Name' : genre_map.keys(), 'Genre_ID': genre_map.values()})
genre_lookup.head()
```

## Creating MySQL tables with a primary key using Python 

Note: You will need to make sure to set the Primary Key for each table that you create with pandas + sqlalchemy that is not a joiner table

Using tconst (or imdb_id) as the primary key 

- To use a string column as our primary key, we will need to provide the SQL datatype schema of our columns when we create the table
  - We will then use our engine to run the command to alter our table and add tconst as the primary key 

### Creating a data type schema for to_sql
- Create a dictionary with each column name as a key and the corresponding SQL datatype as the value 

Note on string columns: 

- For columns with 'object' dtype, you should use:
  - the CHAR type if the column will become your primary key and has a fixed length
  - otherwise, use the VARCHAR data type
  - BOTH CHAR/VARCHAR will need the maximum string length to use for that column 

```python
# get max string length 
# max_str_len = df['column'].fillna('').map(len).max()
```

```python
from sqlalchemy.types import *
```

```python
# genres table
rename_map = {'Genre_Name': 'genre_name', 'Genre_ID': 'genre_id'}
genre_lookup = genre_lookup.rename(rename_map, axis = 1)
genre_lookup.head()
```

```python
max_str_len = genre_lookup['genre_name'].fillna('').map(len).max()
max_str_len
```

```python
schema_dict = {'genre_name': CHAR(max_str_len+1),
              'genre_id': INTEGER()}
```

```python
genre_lookup.to_sql('genres', conn, dtype = schema_dict, if_exists = 'replace', index = False)
```

```python
engine.execute('ALTER TABLE genres ADD PRIMARY KEY (`genre_id`);')
```

```python
tmdb_data = tmdb_data.drop_duplicates()
tmdb_data.duplicated().sum()
```

```python
# Save to sql with dtype and index=False
tmdb_data.to_sql('tmdb_data',conn, dtype=schema_dict2, if_exists = 'replace', index=False)
```
